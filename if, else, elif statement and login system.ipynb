### if / else / elif statement

```python
handsome='true'
good_salary='true'
if handsome=='true' and good_salary=='true':
    print("pyari or khubsorat model sa shadi ho ge")
elif handsome!='true' and good_salary=='true':
    print("pyari or khubsorat larki sa shadi ho ge")
elif handsome=='true' and good_salary!='true':
    print("larki sa shadi ho he jaye ge")

else: 
    print("Allah ko jo manzoor")
```
